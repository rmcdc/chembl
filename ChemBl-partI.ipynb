```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

### Dopamine receptor D4 is linked to many neurological and psychiatric conditions including schizophrenia and bipolar disorder, ADHD, addictive behaviors, Parkinson's disease, and eating disorders such as anorexia nervosa

**Other targets that you can use from ChemBL will be provided to you as dataset_chembl.csv**

```python
# Target search 
target = new_client.target
target_query = target.search('Dopamine D4 receptor')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Selecting target
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
#Filtering target by IC50
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
#Saving as dataframe 
df = pd.DataFrame.from_dict(res)
```

```python
# Reading dataframe
df
```

```python
# Data shape
df.shape
```

```python
#Saving to csv format
df.to_csv('dopamine_data_raw.csv', index=False)
```

```python
# Handling missing values
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
# Step to remove duplicates
len(df2.canonical_smiles.unique())
```

```python
# Removing duplicates
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
#Selecting rows from dataframe and saving as another dataframe
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

```python
#Saving as csv
df3.to_csv('dopamine_bioactivity_data_preprocessed.csv', index=False)
```

```python
# Reading the CSV
df4 = pd.read_csv('dopamine_bioactivity_data_preprocessed.csv')
```

```python
# Printing df4
df4
```

```python
df4.info()
```

#Labeling compounds as either being active, inactive or intermediate

```python
# Creating an emty list and appending values
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
df4.to_csv('df4.csv', index=False)
```

```python
# Adding column class 
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
#Saving CSV
df5.to_csv('dopamine_bioactivity_data_curated.csv', index=False)
```
